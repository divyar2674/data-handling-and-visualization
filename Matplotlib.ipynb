```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[5,6,7,8]
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line graph")
plt.plot(x,y)
plt.show()
```

```python
#pie chart
x=[20,30,40,50]
y=["bat","ball","bag","jar"]
A=[0.2,0.3,0,0]
plt.pie(x,labels=y,explode=A)
plt.show()
```

```python
X=[1,2,3,4]
Y=[7,8,9,10]
plt.xlabel("X AXIS")
z=['r','blue','pink','green']
plt.bar(X,Y,color=z)
plt.show()
```

```python
x=[2,4,6,8]
y=[1,3,5,7]
z=['r','blue','pink','green']
plt.barh(x,y,color=z)
plt.show()
```

```python
x=[2,4,6,8]
y=[1,3,5,7]
z=['r','blue','pink','green']
sizes=[2000,300,400,500]
plt.scatter(x,y,color=z,marker="*",edgecolors="green",s=sizes)
plt.show()
```
