```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
```

```python
text=("A powerful fox known as the Nine-Tails attacks the ninja village Konoha. In response, the leader of Konoha -the Fourth Hokage -seals the fox inside his newborn son Naruto Uzumaki at the cost of his life. As a child, Naruto is isolated from the Konoha community, who regards him as if he was the Nine-Tails itself.")
wordcloud=WordCloud(width=800,height=400,background_color="black").generate(text)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.title("word cloud")
plt.show()
word=text.split()
word_c=Counter(word)
plt.figure(figsize=(10,5))
plt.bar(word_c.keys(),word_c.values())
plt.title("histogram of word frequency")
plt.xlabel("words")
plt.ylabel("frequency")
plt.xticks(rotation=90)
plt.show()
```
