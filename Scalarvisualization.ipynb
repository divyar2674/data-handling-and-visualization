```python
#colour maps
import pandas as p
import numpy as n
import matplotlib.pyplot as plt
data=p.DataFrame ({"x":n.random.randn(100),"y":n.random.randn(100),"value":n.random.randn(100)})
cmap="viridis"
alpha=1
#sactter plot
plt.figure(figsize=(6,6))
plt.scatter(data["x"],data["y"],c=data["value"],cmap=cmap,alpha=alpha)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot with column")
plt.colorbar(label="value")
plt.show()
```

```python
#contouring
data=p.DataFrame(n.random.rand(10,10))
plt.figure(figsize=(8,5))
plt.contour(data.values)
plt.show()
```

```python
#height plot
from mpl_toolkits.mplot3d import Axes3D
data=p.DataFrame(n.random.rand(10,10))
x=n.arange(data.shape[0])
y=n.arange(data.shape[1])
X,Y=n.meshgrid(x,y)
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(111,projection="3d")
ax.plot_surface(X,Y,data.values,cmap='viridis')
plt.show()
```
