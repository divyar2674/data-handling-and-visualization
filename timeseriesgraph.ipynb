```python
import pandas as pd
import matplotlib.pyplot as pt
```

```python
data={"date":pd.to_datetime(["2022-01-01","2022-02-02","2022-03-26","2022-04-27","2022-07-28"]),"value":[5,10,15,20,25]};
df=pd.DataFrame(data)
```

```python
df.set_index("date",inplace=True)
pt.figure(figsize=(10,6))
pt.plot(df.index,df["value"],marker='X')
pt.xlabel("date")
pt.ylabel("value")
pt.title("time series graph")
pt.grid(True)
pt.show()
```
